```julia
using Plots
include("../ensemble.jl")
include("../exact/random_regular.jl")
import ProgressMeter; ProgressMeter.ijulia_behavior(:clear);
```

## β=1.0
This is above the ferromagnetic transition @ $β=\tanh^{-1}\left(\frac{1}{J(k-1)}\right)\approx 0.5$

```julia
β = 1.0
T = 25     # final time
ε = 5e-2  # truncation error for svd   
pᵢ⁰ = [0.75, 0.25]   # initial condition: magnetization = +0.5
ϕᵢ = [ones(2) for _ in 1:T]   # observations (none)
wᵢ = fill( GlauberFactor(ones(2), 0.0, β), T )   # transition probabilities: J=1, h=0

A = rand_mpem2(2, T, d=1)   # initialize with bonds of dimension d
Δs = zeros(0);
```

Run on infinite 3-regular graph

```julia
A, iters, Δs = iterate_rs_deg3(A, pᵢ⁰, ϕᵢ; ε, maxiter=100, tol=1e-3, Δs)
println("### Computing magnetization...\n")
mag = magnetization_rs_deg3(A, pᵢ⁰, ϕᵢ; ε)
```

Compute RS equilibrium prediction

```julia
k = 3
J = 1.0
m_eq = rs_magnetization_fixedpoints(β, J, k) 
```

```julia
p1 = plot(0:T, mag, title="T=$T, β=$β, ε=$ε", xlabel="time",
    ylabel="magnetiz", m=:o, label="", titlefontsize=12)
hline!(p1, [m_eq[3]], ls=:dash, label="RS @equilibrium")
p2 = plot(1:iters, Δs[1:iters], xlabel="time", ylabel="convergence error", 
    m=:o, label="")
p3 = plot(p1, p2, layout=(2,1), size=(500,500))
```
