```julia
using MatrixProductBP, MatrixProductBP.Models
using JLD2, SparseArrays, PyPlot, Statistics
include("/home/fedflorio/master_thesis/Utilities/roc.jl");
```

```julia
lambda = 0.05
rho = 0.05
seeds = [1 2 3 4 6 8 9 10 12 13]
steps = [1 2 3 4 5 6 7 8]
nsnaps = 100
;
```

```julia
neigs = load("/home/fedflorio/master_thesis/MatrixProductBP.jl/notebooks/sis_inference_data/graphs/ER_add_neigs.jld2")["neigs"]

roc_mean = []
roc_std = []

for step in steps
    areas = []
    for seed in seeds
        path = "./simulations/ER_add_lambda_00$(Int(lambda*1000))_rho_00$(Int(rho*1000))_step$(step)_nobs$(nsnaps)_seed$(seed).jld2"
        D = load(path)
        位_true = D["位"]
        data = D["data"]

        xplot, yplot, area = roccurve_incomplete(data.位, 位_true, neigs)
        push!(areas, area)
    end

    push!(roc_mean, mean(areas))
    push!(roc_std, std(areas))
end
```

```julia
errorbar(steps, roc_mean, yerr=roc_std)
plot(steps, roc_mean)
gcf()
```
